```python
data = pd.read_csv("http://archive.ics.uci.edu/ml/machine-learning-databases/auto-mpg/auto-mpg.data-original", delim_whitespace = True, \
 header=None, names = ['mpg', 'cylinders', 'displacement', 'horsepower', 'weight', 'acceleration', 'model', 'origin', 'car_name'])
```

```python
data.to_csv('carmpg.csv',index = None)
data.columns
```

```python
data['type_car'] = data.car_name.str.split().str.get(0)
```

```python
test = data.groupby(['model'])
```

```python
data_ = test['type_car'].count()
data_
```

```python
sns.set_style("whitegrid")
sns.lineplot(data =data_,legend = 'auto',color= 'orange',marker = 'o').set(title = 'A number of new cars introduced each year',xlabel = 'model',ylabel = 'count')

ax = plt.subplot(111)  
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
```
